# Case Study 8: Stock Performance

Suppose that you have some cash burning a hole in your pocket and want to invest in the stock market. This notebook is designed to provide a tool for analysing a few stocks at a time (based on their market performance and trading volume). As an example, we will present an analysis of 11 stocks.

```python
import pandas as pd
from pandas_datareader import data

from plotnine import *

from bokeh.palettes import Category20
from bokeh.plotting import figure, output_file, output_notebook, show, ColumnDataSource
from bokeh.models import HoverTool, Label
from bokeh.models.widgets import Panel, Tabs
from bokeh.layouts import gridplot

import math

from scipy import stats
```

```python
output_notebook()
```

```python
tickers = ["CXW", "F", "GM", "JCP", "KR", "WDC", "NKE","T", "WDAY", "WFC", "WMT"]
start_date = '2012-11-01'
end_date = '2017-11-01'
data_source = 'yahoo'

INVESTMENT = 25000.0
INVESTMENT_PER_TICKER = INVESTMENT / len(tickers)
```

```python
# Load in the financial data
panel_data = data.DataReader(tickers, data_source, start_date, end_date)
```

```python
# Separate the data in two DataFrames, one for volumne
# and one for adjusted closing price. 
close = panel_data['Adj Close'].reset_index()
volume = panel_data['Volume'].reset_index()

# Make the data "tidy"
tidy_volume = pd.melt(volume.reset_index(),
                      id_vars='Date',
                      value_name='Volume',
                      value_vars=tickers,
                      var_name='Ticker')

tidy_close = pd.melt(close.reset_index(),
                      id_vars='Date',
                      value_name='Value',
                      value_vars=tickers,
                      var_name='Ticker')

# Combine the adjusted closing data with the volume data
combined = pd.merge(tidy_close, tidy_volume, on=['Ticker','Date'])

# Add columns to tell us how many shares we started out with
# and how much money we've made/lost.
opening_vals = {ticker:close[ticker].iloc[-1] for ticker in tickers}
combined['opening_val'] = combined['Ticker'].apply(lambda x: opening_vals[x])
combined['num_shares'] = combined['Ticker'].apply(lambda x: math.floor(INVESTMENT_PER_TICKER / opening_vals[x]))
combined['revenue'] = combined['num_shares']*(combined['Value'] - combined['opening_val'])

# Create a DataFrame that tells us our total revenue over time. 
totalled = pd.DataFrame(combined.groupby(['Date'])['revenue'].sum()).reset_index()
```

## Stock Performance

The following presents the price of 11 stocks over the past five years. Click on the ticker symbols to hide/show the corresponding line on the graph. Then, choose the second or third tab to see revenue information. Revenue is here defined as the price of the stock times the numbers of stocks minus the initial investment. 

```python
# Build stock performance plot
p1 = figure(plot_width=900, plot_height=600, x_axis_type="datetime")
p1.title.text = '5 Year Stock Performance'
p1.yaxis.axis_label = 'Stock Price ($)'
for name, color in zip(tickers, Category20[11]):
    p1.line(close['Date'], close[name], line_width=2, color=color, alpha=0.8, muted_color=color, muted_alpha=0.2, legend=name)
p1.legend.location = "top_left"
p1.legend.click_policy="mute"
tab1 = Panel(child=p1, title='Stock Performance')

# Build revenue plot
p2 = figure(plot_width=900, plot_height=600, x_axis_type="datetime")
p2.title.text = '5 Year Revenue'
p2.yaxis.axis_label = 'Revenue ($)'
for name, color in zip(tickers, Category20[11]):
    df = combined[combined['Ticker'] == name].copy()
    p2.line(df['Date'], df['revenue'], line_width=2, color=color, alpha=0.8, muted_color=color, muted_alpha=0.2, legend=name)
p2.legend.location = "top_left"
p2.legend.click_policy="mute"
tab2 = Panel(child=p2, title='Revenue')

# Build total revenue plot
p3 = figure(plot_width=900, plot_height=600, x_axis_type='datetime')
p3.title.text = '5 Year Revenue from All Stocks'
p3.yaxis.axis_label = 'Revenue ($)'
p3.line(totalled['Date'], totalled['revenue'], line_width=2)
tab3 = Panel(child=p3, title='Total Revenue')

tabs = Tabs(tabs=[tab1, tab2, tab3])

show(tabs)
```

## Volume

One important consideration when buy stocks is the stock trading volume (the number of times that a particular share changes hands in the course of a day). Directly below is a depiction of the volume of each stock. Following is a plot comparing the value stocks versus their volume. A downward trend on this plot would indicate suggest that stock that get traded more often would tend to be of lesser value. 

```python
# Construct plots about volume
plots = []
for name, color in zip(tickers, Category20[len(tickers)]): 
    p = figure(x_axis_type="datetime", y_range=(0,15e7))
    p.title.text = '{} Volume'.format(name)
    p.yaxis.axis_label = 'Trade Volume'
    p.left[0].formatter.use_scientific = False
    p.line(volume['Date'], volume[name], line_width=2, color=color)
    plots.append(p)  
grid = gridplot(plots,  ncols=4, plot_width=220, plot_height=250)
show(grid)
```
