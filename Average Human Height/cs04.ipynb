```python
import pandas as pd
from plotnine import *
import plotnine as pln
```

```python
"""
Read in the 6 Data Sets. The .sav and .dbf files were troublesome to import
directly with python, so they were converted to .csv files elsewhere. 
"""
heightXL = pd.read_excel("https://byuistats.github.io/M335/data/heights/Height.xlsx", skiprows=2)
conscript = pd.read_stata("germanconscr.dta")
prison = pd.read_stata("germanprison.dta")
soldiers = pd.read_csv("soldier_heights.csv")
wage_height = pd.read_csv("wage_height.csv")
nat_survey = pd.read_csv('nat_survey.csv')
```

```python
# Gather the columns of the dataframe into 'tidy' columns

heightXL = heightXL.melt(id_vars=['Code', 'Continent, Region, Country'], 
                         var_name='year', value_name='height')
heightXL.columns = ['Code', 'Country', 'year', 'height']
```

```python
height1 = conscript.drop(['gebger', 'age', 'co'], axis=1)
height1['study_id'] = 'CONS'
```

```python
height2 = prison.drop(['age','co'], axis=1)
height2['study_id'] = 'PRIS'
```

```python
height3 = soldiers[['GEBJZ','CMETER']].copy()
height3.columns = ['bdec', 'height']
height3['bdec'] = height3['bdec'] * 10
height3['study_id'] = 'SOLD'
```

```python
height4 = wage_height[['height']].copy()
height4['bdec'] = 1950
height4['height'] = height4['height'] * 2.54
height4['study_id'] = 'WAGE'
```

```python
height5 = nat_survey[['DOBY', 'RT216F', 'RT216I']].copy()
height5['bdec'] = ((height5['DOBY'] + 1900) // 10) * 10
height5['height'] = (height5['RT216F'] * 12 + height5['RT216I']) * 2.54
height5.drop(['RT216F', 'RT216I', 'DOBY'], axis=1, inplace=True)
height5['study_id'] = 'NATS'
```

```python
# Combine the dataframe that we formatted above. 

all_heights = pd.concat([height1, height2, height3, height4, height5], ignore_index=True)
```

```python
pln.options.figure_size = (12, 4.8)
(ggplot() +
  geom_point(data = heightXL, mapping=aes(x='year', y='height')) +
  geom_point(data = heightXL[heightXL['Code'] == 276], mapping=aes(x='year', y='height'), color = "red") +
  scale_x_discrete(breaks = range(1800, 2000, 10)) +
  labs(title = "Worldwide Height Estimates",
       y = "Height (cm)",
       x= "Year of Birth") +
  annotate("text", x = 190, y = 181, label = "Germany", color = "red") +
  theme_bw())
```

This plot show estimates at the average height of men in many countries across the world. Germany has been highlighted as an example of a country whose averge height seems to have increased over the years.

```python
century_finder = lambda x: ('18th Century' if 1700 <= x < 1800 else 
                            '19th Century' if 1800 <= x < 1900 else 
                            '20th Century')
all_heights['century'] = all_heights['bdec'].apply(century_finder)
all_heights = all_heights[(all_heights['height'] > 0) & (all_heights['height'] < 272)]
all_heights_grouped = all_heights.groupby(['century', 'bdec']).mean().reset_index()

(ggplot(all_heights_grouped, aes(x='bdec', y='height')) +
 geom_point() +
 geom_line(color='blue') + 
 facet_grid('. ~ century', scales='free_x') + 
 labs(title = "Average Height by Century",
    y = "Height (cm)",
    x = "Year of Birth") +
 theme_bw())
```

The above plot shows the average height of males from 5 different studies over 3 centuries. The dramatic dip in the 19 century throws some doubt over the validity of the data.
